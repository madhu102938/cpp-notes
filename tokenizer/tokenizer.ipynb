```python
def get_stats(ids) -> dict[tuple[int, int], int]:
    stats = {}
    for (i, j) in zip(ids, ids[1:]):
        stats[(i, j)] = stats.get((i,j), 0) + 1
    
    return stats

text_encode = list(text.encode("utf-8"))
stats = get_stats(text_encode)
sorted([(v, k) for k, v in stats.items()], reverse=True)
```

```python
def change(_text: list[int], a: tuple[int, int], new: int) -> list[int]:
    n:int = len(_text)
    i:int = 0
    newids = []
    while i < n:
        if i+1 < n and (_text[i], _text[i+1]) == a:
            newids.append(new)
            i += 2
        else:
            newids.append(_text[i])
            i += 1
    return newids
```

```python
vocab = {idx : bytes([idx]) for idx in range(256)}

for (i, j), v in merge.items():
    vocab[v] = vocab[i] + vocab[j]
    
def decode(_text:list[int]) -> str:
    decode_text = b"".join(vocab[i] for i in _text)
    return decode_text.decode("utf-8", errors='replace')
```

```python
all_merges = [(a,b) for (a,b),c in merge.items()]
all_merges
```

```python
def encode(_text:str):
    level_1 = list(_text.encode('utf-8'))
    while len(level_1) > 1:
        to_change:tuple[int, int] = (-1, -1)
        mini:int = int(2e9)
        for i, j in zip(level_1, level_1[1:]):
            if all_merges.count((i, j)) > 0:
                temp = all_merges.index((i,j))
                if temp < mini:
                    mini = temp
                    to_change = (i, j)
        if to_change[0] != -1:
            # print (to_change, merge[to_change])
            level_1 = change(level_1, to_change, merge[to_change])
        else:
            break
    
    return level_1
```
